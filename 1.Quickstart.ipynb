# Начало работы с vim

Редактор vim работает в 3 основных режимах:

1. Нормальный - режим навигации по тексту.  Активен при начале работы, активизируется клавишей `Esc`.
2. Режим вставки. Позволяет вводить текст. Активизируется клавишей `i`.
3. Командный. Активизируется клавишей `:` из режима вставки. Деактивизируется нажатием `Esc` или вводом команды.

## Как выйти из vim

Очень просто. В командном режиме набрать `:q!`. Восклицательный знак нужен, чтобы выйти без сохранения.

## Как сохранить написанное в vim

Командой `:w`. Если надо сохранить и сразу выйти, это выполняется командой `:wq`.

# Основы vim

## Основные навигационные команды

1.  `h` - курсор влево
2.  `j` - курсор вниз
3.  `k` - курсор вверх
4.  `l` - курсор вправо
5.  `w` - переход на слово вперед (разделителем слов считается любой знак препинания)
6.  `W` - переход на слово вперед (разделителем слов считается пробел)
7.  `b` - переход на слово назад (разделителем слов считается любой знак препинания)
8.  `W` - переход на слово назад (разделителем слов считается пробел)
9.  `Ctrl+f` - Page Up
10. `Ctrl+b` - Page Down
11. `gg` - перейти в начало файла
12. `G` - перейти в конец файла
13. `:set number` - отображать номера строк
14. `42:<Enter>` - перейти на строку 42
15. `z+<Enter>` - перейти к началу следующей строки
16. `Z+<Enter>` - перейти вниз с сохранением позиции курсора
17. `0` - перейти к началу строки
18. `^` - перейти к началу абзаца
19. `$` - перейти к концу строки (перед последним символом)
20. `A` - перейти к концу строки после после последнего символа (vim переходит в режим вставки)
21. `Ctrl+g` - показать информацию о местонахождении в файле

## Удаление символов
1.  `u` - отменить последнее действие
2.  `2u` - отменить 2 последних действия
3.  `x` - удалить следующий за курсором символ
4.  `X` - удалить символ перед курсором
5.  `dw` - удалить слово после курсора
6.  `dd` - удалить строку
7.  `5dd` - удалить 5 строк
8.  `d0` - удалить символы от начала строки до текущего
9.  `d$` - удалить символы от текущего до конца строки
10. `.` - повторить последнюю команду

## Буфер обмена

Описанные ранее команды удаления - `x`, `dd` и т.д. на самом деле не удаляют текст, а вырезают его и помещают в буфер обмена. Буфер обмена vim состоит из регистров, в каждом из которых можно хранить отдельный объект. При нажатии `dd` используется регистр по-умолчанию (безымянный регистр). Кроме безымянного регистра есть нумерованные регистры - их 9 штук. Нумерация начинаятся с 0 и заканчивается 9. Для обращения к нумерованному регистру исполльзуется знак `"` и номер регистра, так для обращения к регистру 4 используется `"4`.

1.  `p` - вставить содержимое безымянного регистра.
2.  `v` - выделить блок текста.
3.  `V` - выделить строку
4.  `y` - копировать (yank) выделенный текст
5.  `yw` - копировать слово
6.  `2yw` - копировать 2 слова
7.  `yy` - копировать строку
8.  `u` - отменить последнюю операцию
9.  `Ctrl+r` - повторить последнюю операцию
10. `:reg` - просмотреть содержимое регистров

## Справочная система

Для доступа к справочной системе используется команда `:help`. Команды справочной системы: 

1. `:q` - выйти из справочной системы. 
2. `:help <cmd>` получить справку по команде `<cmd>`.
3. `Ctrl+o` - перейти к предыдущему разделу справочной системы (как кнопка Назад в веб-браузере).
4. `Ctrl+i` - перейти к следующему разделу справочной системы.
5. `Ctrl+]` - перейти по ссылке в справочной системе
6. `Ctrl+w+w` - переход между окном справки и редактора
7. `Ctrl+d` - при работе в командной строке это сочетание активирует command completion, т.е. по префиксу выдает возможные команды.
